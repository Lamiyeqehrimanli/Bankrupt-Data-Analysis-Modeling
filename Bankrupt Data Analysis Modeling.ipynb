```python
df.drop(columns = ['net_income_flag', 'liability_assets_flag'], axis=1, inplace=True)
```

```python
X = df.drop('bankrupt?', axis=1)
y = df['bankrupt?']
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,y, test_size =30, random_state=101, shuffle=True, stratify=y)
```

```python
from sklearn.preprocessing import RobustScaler
scaler = RobustScaler()
scaler =scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled.shape, X_test_scaled.shape, y_train.shape, y_test.shape
```

```python
X_train = copy.deepcopy(X_train_scaled)
X_test = copy.deepcopy(X_test_scaled)
```

```python
#Bagging Classifier
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier

bagging_DT_model = BaggingClassifier(
    estimator=DecisionTreeClassifier(random_state=42),
    n_estimators=150,
    max_samples=0.5,
    max_features=8,
    bootstrap=True,
    bootstrap_features=True,
    n_jobs=-1,
    random_state=101
)
```

```python
bagging_DT_model.fit(X_train,y_train)
```

```python
 bagging_DT_model.score(X_test,y_test)
```

```python
bagging_DT_model_pred = bagging_DT_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,bagging_DT_model_pred)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, classification_report
print("Confusion Matrix:\n", confusion_matrix(y_pred=bagging_DT_model_pred, y_true=y_test))

print(classification_report(y_true=y_test, y_pred=bagging_DT_model_pred))

print("\n Accuracy score:", accuracy_score(y_test, bagging_DT_model_pred),
     "\n Precision score:", precision_score(y_test, bagging_DT_model_pred),
      "\n Recall score :", recall_score(y_test, bagging_DT_model_pred),
      "\n F1 score:", f1_score(y_test, bagging_DT_model_pred))
import warnings
warnings.filterwarnings("ignore")
```
